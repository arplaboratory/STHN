```python
from PIL import Image
import matplotlib.pyplot as plt
import yaml
import os
import numpy as np
import cv2

folder_config_path = './global_pipeline/folder_config.yml'
datasets_folder = './datasets/'
Image.MAX_IMAGE_PIXELS = None
```

```python
# Load folder config
with open(folder_config_path, 'r') as f:
    folder_config = yaml.safe_load(f)
print(folder_config)
```

```python
# Load images
valid_region_uav = [270, 1500, 2700, 9400] # Top left bottom right
satellite_image_1 = np.array(Image.open(os.path.join(datasets_folder, folder_config['satellite']['name'], folder_config['satellite']['maps'][0])))[valid_region_uav[0]:valid_region_uav[2], valid_region_uav[1]:valid_region_uav[3], :]
thermal_image_1 = np.array(Image.open(os.path.join(datasets_folder, folder_config['thermalmapping']['name'], folder_config['thermalmapping']['maps'][0])))[valid_region_uav[0]:valid_region_uav[2], valid_region_uav[1]:valid_region_uav[3], :]
```

```python
# Show
plt.figure(dpi=200)
plt.imshow(satellite_image_1)
plt.figure(dpi=200)
plt.imshow(thermal_image_1)
```
